```python
import seaborn as sns

#import dataset from seaborn
titanic = sns.load_dataset('titanic')

titanic
```

```python
titanic.info()
# ada missing data
# we need to handle the missing data accordingly
```

```python
#handle null value

#fill using interpolation
titanic["age"] = titanic.age.interpolate()
# if the number tend to gown down, then go down, if tend to increase then increase
#"pads" copy previous "nearest": choose nearby
titanic["age"][300:310]
```

```python
#fill specific value
titanic["embark_town"] = titanic.embark_town.fillna("Southampton")
titanic.info()
```

```python
# handle missing value by row or col. The more data remained then save it 
# dropped_titanic = titanic.dropna(axis=0) #drop the row
# dropped_titanic.info()
# SCARYYY 
```

```python
dropped_titanic = titanic.dropna(axis=1) #drop the col
dropped_titanic.info()
# embark没了
```

```python
titanic
```

```python
dropped_titanic
```

```python
grouped_titanic = dropped_titanic.groupby("survived")
# 1 for survived,0 for not survived
for grp, content in grouped_titanic:
    print(grp,content)
# pclass= class, survived=alive, adult_male= sex+child
```

```python
import pandas as pd
result_list=[]
for grp,content in grouped_titanic:
#dataframe = dictinary in list [{},{},{}], dict representation the row = lists of dict
    result_dict = {} # create a new one every time it run the loop

    
    result_dict["survival_status"] = grp
    # inner-for is content
    for who in content["who"]:
        if who not in result_dict.keys():
            #first value
            result_dict[who] = 1
        else:
            result_dict[who] += 1

    result_list.append(result_dict)

result_df = pd.DataFrame(result_list)
result_df
        
```

```python
for rdict in result_list:
    for k, v in rdict.items():
        if k == "survival_status" and v == 0:
            rdict["survival_status"] = "Not Survived"
        elif k == "survival_status" and v == 1:
            rdict["survival_status"] = "Survived"

result_df = pd.DataFrame(result_list)
result_df

# define above also ok, when we pull the data, we change the name 
```
