```python
tips = sns.load_dataset("tips")
tips.plot.pie(x="sex",y="tip")
#it takes data 1 by 1 ,look crazy
```

```python
group_tips = tips.groupby("sex")

tip_stats = []
for grp, info in group_tips:
    tip_dict = {}
    
    tip_dict["Gender"] = grp
    tip_dict["Tips"] = round(info["tip"].mean(), 2)
    tip_dict["Total Bill"] = round(info["total_bill"].mean(), 2)
    tip_dict["Tips_Percent"] = round(info["tip"].mean() / info["total_bill"].mean()*100, 2)
    
    tip_stats.append(tip_dict)

# Display the results
tip_df = pd.DataFrame(tip_stats)
tip_df
```

```python
tip_df.plot.pie(x= "Gender",y="Tips_Percent")
labels = tip_df.Gender.unique
plt.legend(labels)
```
